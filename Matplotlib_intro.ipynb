# Bar chart

```python
data2 = [5. , 25. , 50. , 20.]
plt.bar(range(len(data2)), data2)
plt.show()
```

# Horizontal Bar Chart 

```python
data2 = [5. ,25. ,50. ,20.]
plt.barh(range(len(data2)),data2)
plt.show()
```

# Error Bar Chart

```python
x9 = np.arange(0,4,0.2)
y9 = np.exp(-x9)
e1 = 0.1 * np.abs(np.random.randn(len(y9)))
plt.errorbar(x9,y9,yerr = e1, fmt = '.-')
plt.show()
```

# Stacked Bar Chart

```python
A = [15. ,30. ,45. ,22.]
B = [15. ,25. ,50. ,20.]
z2 = range(4)
plt.bar(z2, A,color = 'b')
plt.bar(z2, B,color = 'r',bottom = A)
plt.show()
```

# Pie Chart

# Adding a legend

```python
x15 = np.arange(1,10)
fig,ax = plt.subplots()
ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)
ax.legend(['Normal','Fast','Slow'])
plt.show()
```

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

ax.legend()
plt.show()

```
