# Graph Representations

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x="total_bill", kde=True)
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size",palette='deep', sizes=(20, 200), alpha=0.5)
plt.title("Scatterplot of Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(data=tips, x="size", y="total_bill", hue="sex",palette='deep', markers="o",ci=None)
plt.title("Lineplot of Total Bill vs Tip")
plt.show()
```

```python
sns.lineplot(data=tips, x="size", y="total_bill", hue="sex",palette='deep', markers="o")
plt.title("Lineplot of Total Bill vs Tip")
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',hue='sex',palette='muted')
plt.title("Barplot of Total Bill vs Day")
plt.show()
```

```python
tips.columns
```

```python
sns.boxplot(data=tips,x='day',y='tip',hue='smoker',palette='Set2')
plt.title("Boxplot of Tip vs Day")
plt.show()
```

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='time',palette='pastel',split=True)
plt.title("Violinplot of bill vs Day")
plt.show()
```

```python
sns.countplot(data=tips,x='day',hue='smoker',palette='dark')
plt.title("Countplot of Day vs Smoker")
plt.show()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title("Regression Plot of Total Bill vs Tip")   
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='blue')
plt.title("Histogram of Total Bill with KDE")
plt.show()
```

```python
sns.pairplot(tips, hue='sex',vars=['total_bill','tip','size'], palette='husl')
plt.title("Pairplot of Numerical Variables by gender",y=1.02)
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='tip',hue='sex',kind='point',palette='bright')
plt.title("Catplot (point) of Tip vs Day")
plt.show()
```

```python
sns.jointplot(data=tips,x='total_bill',y='tip',kind='scatter',color='purple',palette='coolwarm')
plt.title("Jointplot of Total Bill vs Tip")
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='smoker',  palette='coolwarm')
plt.suptitle("Jointplot: Total Bill vs Tip", y=1.02)
plt.show()
```

```python
sns.FacetGrid(data=tips,col='time',row='smoker',margin_titles=True).map(sns.histplot, 'total_bill', kde=True, color='purple').add_legend()
plt.suptitle("FacetGrid of Total Bill by Time and Smoker", y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(tips, col='time', row='smoker', margin_titles=True).map(sns.histplot, 'total_bill', bins=20, kde=True).add_legend()
g
```

```python
sns.stripplot(data=tips, x='day', y='tip', hue='sex', jitter=True, palette='Set1')
plt.title("Stripplot of Tip by day and gender")
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill',hue='sex', palette='tab10', fill=True, alpha=0.5)
plt.title("KDE Plot of Total Bill density by gender")
plt.show()
```
