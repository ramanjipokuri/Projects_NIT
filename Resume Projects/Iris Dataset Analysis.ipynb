# KDE Plot

```python
sub=iris[iris['Species']=='Iris-setosa']
sns.kdeplot(x=sub['SepalLengthCm'],y=sub['SepalWidthCm'],fill=True ,cmap='plasma',shade=True,shade_lowest=False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length CM')
plt.ylabel('Sepal Width CM')
plt.show()
```

```python
sub = iris[iris['Species'] == 'Iris-setosa']

# Create KDE plot
plt.figure(figsize=(8,6))
sns.kdeplot(x=sub['SepalLengthCm'], y=sub['SepalWidthCm'], fill=True, cmap='plasma')

# Titles and labels
plt.title('Iris-setosa Sepal Length vs Width KDE Plot')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# Stacked Histogram

```python
iris['Species'] = iris['Species'].astype('category')
```

```python
iris.head()
```

```python
list1=list()
mylabels=list()
for gen in iris.Species.cat.categories:
    list1.append(iris[iris.Species==gen].SepalLengthCm)
    mylabels.append(gen)
    
h=plt.hist(list1,bins=30,stacked=True,rwidth=1,label=mylabels)
plt.legend()
plt.show()
```
