# Introduction to Seaborn

```python
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
```

```python
df.head()
```

```python
vis1 = sns.histplot(df["InternetUsers"])
plt.show()
```

```python
vis1 = sns.histplot(df["InternetUsers"], bins=10) 
plt.show()
```

```python
vis2 = sns.boxplot(data = df, x="IncomeGroup", y='BirthRate')
plt.show()
```

```python
vis3 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate', fit_reg = False)
plt.show()
```

```python
vis4 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate')
plt.show()
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup')
plt.show()
```

```python
vis5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup', height = 10)
plt.show()
```
