# Univariate

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
pd=pd.DataFrame({"Age":[20,21,25,30,18,40,35,40,35,12]})
```

```python
sns.histplot(pd['Age'],bins=5,kde=True)
plt.title("Age distribution")
plt.show()
```

```python
sns.boxplot(y=pd['Age'])
plt.title("Age distribution")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.DataFrame({"Category":['A','B','C','D','E'],"Count":[10,20,18,30,15]})
```

```python
sns.barplot(x=df['Category'],y=df['Count'])
plt.title("Catefgory Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({"Category": ["A", "B", "C", "D"], "Values": [10, 15, 7, 12]})

df.plot.pie(y="Values", labels=df["Category"], autopct="%1.1f%%", legend=False)

plt.title("Category Distribution")
plt.ylabel("")
plt.show()
```

# Bivariate

```python
df=pd.DataFrame({"Exp":[1,2,4,7,9,10,5,7],
                 "Salary":[30000,60000,120000,170000,200000,270000,100000,190000]
                })
```

```python
sns.scatterplot(x=df['Exp'],y=df['Salary'])
plt.title("Exp vs Salary")
plt.show()
```

```python
sns.lineplot(x=df['Exp'],y=df['Salary'],marker='d')
plt.title("Exp and salary")
plt.show()
```

```python
sns.boxplot(x=df['Exp'],y=df['Salary'])
plt.title("salary distribution by exp")
plt.show()
```

```python
dd = pd.DataFrame({
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "Salary": [50000, 55000, 60000, 58000, 62000, 57000]
})
sns.boxplot(x=dd["Gender"], y=dd["Salary"])
plt.title("Salary Distribution by Gender")
plt.show()
```

```python
sns.violinplot(x=df['Exp'],y=df['Salary'])
plt.title("salary distribution (violinplot)")
plt.show()
```

```python
df=pd.DataFrame({"Age":[20,22,24,30,40,43,35,37],
                 "Exp":[1,2,4,7,9,10,5,7],
                 "Salary":[30000,60000,120000,170000,200000,270000,100000,190000]
                })
```

```python
plt.figure(figsize=(8,7))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',linewidths=0.5)
plt.title("correlation heat map")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
pd=pd.DataFrame({"Age":[20,21,25,30,18,40,35,40,35,12]})
sns.countplot(x=pd['Age'],palette="coolwarm")
plt.show()
```
